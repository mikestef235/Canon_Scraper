## Determine if Listing Contains a Camera

```python
#Method 1. Find if title contains a camera model from the elig_cams list

#Make a list of the camera model numbers
cams = elig_cams.columns[1:].tolist()

#Add in the Canon 6D and Rebel T5i for testing
cams.extend(['Canon 6D', "Canon T5i"])

#Add model cams to the set
modelNums = set()
for cam in cams:
    cam = cam[6:]
    modelNums.add(cam)

#Create a copy of the sold items data
Sold_Cams = sold_items.copy(deep=True)

#Creating a set for each item title in sold items
sold_model = []
for item in sold_items.Title:
    string = item
    word_set = set(string.split(' '))
    match = list(word_set & modelNums) #Finding if a string in the item title matches the camera model bank
    if len(match) != 0:
        sold_model.append(match[0]) #Might need to fix if there is more than one model present
    else:
        sold_model.append('N/A')
        
#Add the sold_model list to the Sold_Cams df
Sold_Cams['Model'] = np.array(sold_model)
```

```python
#Method 2. Checking if title column contains body, body-only (meaning it doesn't come with a lens)
searchfor = ['body', 'body-only']
body_only_cams = sold_items.Title.str.lower().str.contains('|'.join(searchfor))
Sold_Cams['Body Only?'] = body_only_cams
```

```python
#Does title have '+', 'with', 'w/', 'and', 'extras
searchfor = ['extras', 'xtras', "w/", 'with', 'and', '\+']
cams_Ex = sold_items.Title.str.lower().str.contains('|'.join(searchfor))
Sold_Cams['Extras?'] = cams_Ex
```

## Examining Descriptions For Accessories

```python
#Determine if Description contains camera bag
searchfor = ['camera bag', 'shoulder bag']
bag = sold_items.Description.str.lower().str.contains('|'.join(searchfor))

#Determine if Description contains filter
searchfor = ['filter', 'filters']
filt = sold_items.Description.str.lower().str.contains('|'.join(searchfor))

#Determine if Description contains memory card
searchfor = ['sd card', 'memory card', 'sd-card', 'memory-card', 'gb']
card = sold_items.Description.str.lower().str.contains('|'.join(searchfor))

#Determine if Description contains batteries
searchfor = ['battery', 'batteries']
batts = sold_items.Description.str.lower().str.contains('|'.join(searchfor))

#Determine if Description contains lens
searchfor = ['lens', 'lenses']
lens = sold_items.Description.str.lower().str.contains('|'.join(searchfor))

#isinstance(batts, pd.Series)
accessoriesDf = pd.DataFrame([bag, filt, card, batts, lens]).T
accessoriesDf.columns=(['Bag?', 'Filter?', 'Memory Card?', 'Battery?', 'Lens?'])
accessoriesDf *= 1
accessoriesDf
```

```python
#Getting the first lens in the listing, extractall in the future
Lens_Spec = (Sold_Cams.Description.str.extract('(\d{2}\-\d{3}mm) | (\d{2}\-\d{2}mm)', expand=True))
Lens_Spec.columns = ['LS1', 'LS2']
Lens_Spec['LS1'].fillna(Lens_Spec['LS2'], inplace = True)
Lens_Spec.drop(['LS2'], axis = 1, inplace = True)
```

```python
#Searching the title for lens spec
Lens_SpecTitle = (Sold_Cams.Title.str.extract('(\d{2}\-\d{3}mm) | (\d{2}\-\d{2}mm)', expand=True))
Lens_SpecTitle.columns = ['LS1', 'LS2']
Lens_SpecTitle['LS1'].fillna(Lens_SpecTitle['LS2'], inplace = True)
Lens_SpecTitle.drop(['LS2'], axis = 1, inplace = True)

#Merging the two 
Lens_Spec.LS1.fillna(Lens_SpecTitle.LS1, inplace=True)
Lens_Spec=pd.DataFrame(Lens_Spec)
```

```python
#Getting the first lens in the listing, extractall in the future
searchfor = ['(f\d.\d)', '(f/\d.\d)', '(\d.\df)', '(f/\d)']
t1 = Sold_Cams.Description.str.lower().str.extract('|'.join(searchfor), expand=True)
t1[0].fillna(t1[3], inplace=True)
t1[0].fillna(t1[2], inplace=True)
t1[0].fillna(t1[1], inplace=True)
t1.drop([1, 2, 3], axis = 1, inplace=True)
```

```python
#Searching the Titles for focal length
searchfor = ['(f\d.\d)', '(f/\d.\d)', '(\d.\df)', '(f/\d)']
t2 = Sold_Cams.Title.str.lower().str.extract('|'.join(searchfor), expand=False)
t2[0].fillna(t2[3], inplace=True)
t2[0].fillna(t2[2], inplace=True)
t2[0].fillna(t2[1], inplace=True)
t2.drop([1, 2, 3], axis = 1, inplace=True)
t2
```

```python
#Merging with the first column
t1[0].fillna(t2[0], inplace=True)
t1.columns = ['Max Aperature']
t1 = pd.DataFrame(t1)
```

```python
lensDf = pd.concat([Lens_Spec, t1], axis=1)
lensDf
```
